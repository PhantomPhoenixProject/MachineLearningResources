# Univariate Linear Regression

## Model Function

```python
import numpy as np
train_x_house_size = np.array([100.0, 120.0, 200.0, 278.5, 310.0])
train_y_house_cost = np.array([500.0, 550.0, 700.0, 760.0, 850.0])

import matplotlib.pyplot as plt
plt.scatter(train_x_house_size, train_y_house_cost, c='r', marker='o')
plt.title("Housing Prices")
plt.xlabel("House size (sq. meters)")
plt.ylabel("Cost (USD)")

plt.show()
```

```python
def predict(input, w, b):
    m = input.shape[0]
    output = np.zeros(m)
    for i in range(m):
        output[i] = (w * input[i]) + b
    return output
#3.0646,  1.5851
w,b=1.5426094964771933,360.8556394623737
output_prediction = predict(train_x_house_size, w,b)

plt.plot(train_x_house_size, output_prediction, c='g', label='our prediction')
plt.scatter(train_x_house_size, train_y_house_cost, c='r', marker='o', label='actual values')
plt.title("Housing Prices")
plt.xlabel("House size (sq. meters)")
plt.ylabel("Cost (USD)")

plt.show()
```

## Cost Function

```python
def cost(x, y, w, b):
    m = x.shape[0]
    res = 0
    for i in range(m):
        res+= ((w*x[i]+b)-y[i])**2
    res = res / (2*m)
    return res

print(cost(train_x_house_size, train_y_house_cost, 1.5426094964771933, 360.8556394623737))
print(cost(train_x_house_size, train_y_house_cost, 2.5, 300))
print(cost(train_x_house_size, train_y_house_cost, 3.5, 400))
print(cost(train_x_house_size, train_y_house_cost, 1.0, 100))
print(cost(train_x_house_size, train_y_house_cost, 0.5, 450))
```
